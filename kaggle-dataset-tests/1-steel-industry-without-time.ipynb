```python
import pandas as pd
from fedot.api.main import Fedot
```

```python
#https://www.kaggle.com/datasets/nimapourmoradi/steel-datase

path = "../datasets/steel-dataset"

df = pd.read_csv(f"{path}/Steel_industry.csv")
df.drop(columns=["Date_Time"], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, train_size=0.8, random_state=42)
```

```python
auto_model = Fedot(
    problem="classification",
    metric=["precision", "accuracy", "roc_auc"],
    preset="best_quality",
    with_tuning=True,
    timeout=2,
    cv_folds=10,
    seed=42,
    n_jobs=1,
    logging_level=10,
    use_pipelines_cache=False,
    use_auto_preprocessing=False
)
```

```python
auto_model.fit(features=train, target="Load_Type")
```

```python
prediction = auto_model.predict(features=test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test.Load_Type, prediction)
```
